```python
import pandas as pd
from pathlib import Path
```

```python
OUTPUT_FILE = Path("/home/agrupa-lab/agrupa/IE_capstones/Omar/outputs/llava_descriptions.csv")

df = pd.read_csv(OUTPUT_FILE)
print(f"Total rows in CSV: {len(df)}")
df.head()
```

```python
df_unique = df.drop_duplicates(subset='cat_no', keep='last').reset_index(drop=True)
print(f"Total unique artworks: {len(df_unique)}")
print(f"Expected: 6266")
print(f"Missing: {6266 - len(df_unique)}")
```

```python
errors = df_unique[df_unique['llava_description'].str.startswith('ERROR', na=False)]
print(f"Error rows: {len(errors)}")
if len(errors) > 0:
    print(errors[['cat_no', 'llava_description']].head(10))
```

```python
# Description length stats
df_unique['desc_length'] = df_unique['llava_description'].str.len()

print("Description length stats:")
print(df_unique['desc_length'].describe())
print(f"\nShortest description:")
print(df_unique.loc[df_unique['desc_length'].idxmin(), 'llava_description'])
print(f"\nLongest description:")
print(df_unique.loc[df_unique['desc_length'].idxmax(), 'llava_description'])
```

```python
print("Coverage by category:")
print(df_unique.groupby(['is_fauna', 'is_religious']).size().reset_index(name='count'))
```
